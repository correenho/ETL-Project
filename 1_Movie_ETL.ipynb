```python
url = f"http://www.omdbapi.com/?apikey={api_key}&T="
link = "https://www.boxofficemojo.com/year/2017/?grossesOption=totalGrosses&sort=openingWeekendGross&ref_=bo_yld__resort#table"
```

```python
print(link)
```

```python
mojo_movies = pd.read_html(link)
```

```python
movies_2017 = mojo_movies[0]
movies_2017 = movies_2017.rename(columns = {"Release":"Title" })
```

```python
movie_list = movies_2017.Title
movies_2017 = movies_2017.drop(columns = ["Genre"])
```

```python
movie_Name = []
Rated = []
Genre = []
Language = []
Awards = []
DVD = []
Box = []
Director = []
Writer =  []
Actors = []
Runtime = []
Plot = []
imdbRating = []
imdbVotes = []
imdbID = []
metascore_list = []
```

```python
for movie in movie_list:
    try:
        response = requests.get(url + movie).json()
        DVD.append(response["DVD"])
        movie_Name.append(response["Title"])
        Rated.append(response["Rated"])
        Genre.append(response["Genre"])
        Language.append(response["Language"])
        Awards.append(response["Awards"])
        Director.append(response["Director"])
        Writer.append(response["Writer"]) 
        Box.append(response["BoxOffice"])
        Actors.append(response["Actors"])
        Runtime.append(response["Runtime"])
        Plot.append(response["Plot"])
        imdbRating.append(response["imdbRating"])
        imdbVotes.append(response["imdbVotes"])
        imdbID.append(response["imdbID"])
        metascore_list.append(response['Metascore'])
    except (KeyError, IndexError):
        print(f'{movie} is Missing Field/result...skipping.')
    except:
        print(f'failed on {movie}')
    
#     time.sleep(1)
```
